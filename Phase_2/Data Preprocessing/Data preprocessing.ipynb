```python
# Import
import pandas as pd

# Load dataset
df = pd.read_csv("credit_card_dataset.csv")

# -------------------------------
# BASIC PREPROCESSING ONLY
# -------------------------------

# Drop useless columns
df = df.drop(['transaction_id', 'customer_id', 'merchant_id'], axis=1)

# Convert time → hour
df['hour'] = df['transaction_time'] // 3600

# Encode categorical columns
df = pd.get_dummies(df, columns=['transaction_type', 'location', 'device_type'])

# Drop original time
df = df.drop('transaction_time', axis=1)

# -------------------------------
# SAVE PROCESSED DATASET
# -------------------------------
df.to_csv("processed_dataset.csv", index=False)

print("Preprocessing Done & Saved!")
print(df.head())
```
