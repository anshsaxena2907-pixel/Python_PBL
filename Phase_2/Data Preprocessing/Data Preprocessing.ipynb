```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("credit_card_dataset.csv")
```

```python
print(df.head())
print(df.info())
```

```python
df.drop(['transaction_id', 'customer_id', 'merchant_id'], axis=1, inplace=True)
```

```python
df = df.dropna()   
```

```python
df['hour'] = df['transaction_time'] // 3600
```

```python
df = pd.get_dummies(df, columns=['transaction_type', 'location', 'device_type'])
```

```python
X = df.drop('fraud', axis=1)
y = df['fraud']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
print("Preprocessing Done!")
print("Shape of Features:", X_scaled.shape)
```
